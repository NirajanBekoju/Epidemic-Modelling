# EPIDEMIC MODELLING with SIR model

S => SUSCEPTIBLE means those who could be potentially infected<br>
I => INFECTED means those who are infected by the epedemic disease<br>
R => REMOVED means those who are either recovered or death<br>

ASSUMPTION IN OUR SIR MODEL
1. The model is valid only for very short period of time.
2. Population remains constant and so no death just recovery from the diseases.
3. Rate of infectives is directly propotional to the contact between the persons.

Using the data obtained from the epidemic modelling, we are going to analyse those data and compare the effect of various factor on epidemic spread over a society. We have considered four factors.<br>
1. Infection Radius
2. Infection Duration
3. No. of people in a society
4. Initial infection percent

```python
# importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Data Name format : "dataf03_120_100_3" <br>
where, <br>
03  => Infection Radius = 0.3<br>
120  => Infection Duration = 120 unit of time<br>
100 => No. of people = 100<br>
3   => Initial infection percent = 3%<br>

```python
""" Read the csv data and plot its SIR Graph and return the dataframe"""
def read_data(name):
    data = pd.read_csv("data/" + name + ".csv", header = None)
    data.columns = ["S", "I", "R"]
    return data

def display_SIR(data):
    plt.figure(figsize=(16, 8))
    plt.plot(data["S"], color = 'g', label="SUSCEPTIBLE")
    plt.plot(data["I"], color = 'r', label="INFECTED")
    plt.plot(data["R"], color = 'b', label="REMOVED")
    plt.title("Variation of Susceptible, Infected and Removed number with time")
    plt.xlabel("Time")
    plt.ylabel("Number of people")
    plt.legend()
    plt.show()

def subplotGraph2D(row, col, dataframe_arr, title_arr, title):
    fig, axs = plt.subplots(row, col, figsize = (16, 8))
    fig.suptitle(title)
    k = 0
    if(row == 1 and col == 2):
        for i in range(col):
            axs[i].plot(dataframe_arr[k]["S"], color="g", label="SUSCEPTIBLE")
            axs[i].plot(dataframe_arr[k]["I"], color="r", label="INFECTED")
            axs[i].plot(dataframe_arr[k]["R"], color="b", label="RECOVERED")
            axs[i].set_title(title_arr[k])
            k+=1
    else:
        for i in range(row):
            for j in range(col):
                axs[i, j].plot(dataframe_arr[k]["S"], color="g", label="SUSCEPTIBLE")
                axs[i, j].plot(dataframe_arr[k]["I"], color="r", label="INFECTED")
                axs[i, j].plot(dataframe_arr[k]["R"], color="b", label="RECOVERED")
                axs[i, j].set_title(title_arr[k])
                k += 1
```

## Affect of Infection Radius on Epidemic Diseases
We have taken the data based on following factor in order to study the affect of infection radius on epidemic diseases spreed.

| Infection Radius | Infection Duration | Population |Initial Infected Percentage|
| :- | :- | :- | :- |
| 0.3 | 120 | 100 | 3% |
| 0.6 | 120 | 100 | 3% |

```python
dataf03_120_100_3 = read_data("dataf03_120_100_3")
dataf06_120_100_3 = read_data("dataf06_120_100_3")
print("Maximum number of infected people at an instant of time : ", dataf03_120_100_3["I"].max())
print("Maximum number of infected people at an instant of time : ", dataf06_120_100_3["I"].max())

title = """Variation of no. of Susceptible, infected and Removed people with time 
120 Infection duration 
100 Populaton 
3% initial infected
"""
title_arr = ["Infection Radius = 0.3", "Infection Radius = 0.6"]
data_arr = [dataf03_120_100_3, dataf06_120_100_3]
subplotGraph2D(1, 2, data_arr, title_arr, title)
```

### Infection Rate Comparison for infection radius = 0.3 and 0.6

```python
plt.figure(figsize=(16, 8))
plt.plot(dataf03_120_100_3["I"], color = 'r', label="0.3 Infection Radius")
plt.plot(dataf06_120_100_3["I"], color = 'b', label="0.6 Infection Radius")
plt.title("Variation of Infection with time \n Infection Radius 0.3 and 0.6")
plt.xlabel("time")
plt.ylabel("No. of infected people")
plt.legend()
plt.show()
```

#### Conclusion from above graph
1. When the infection radius increases, rate of tranmission of the diseases is very high.
2. For high infection radius, higher is the number of infected people at an instant of a time.

```python
dataf03_60_100_3 = read_data("dataf03_60_100_3")
dataf03_120_100_3 = read_data("dataf03_120_100_3")
dataf03_180_100_3 = read_data("dataf03_180_100_3")
dataf03_240_100_3 = read_data("dataf03_240_100_3")


title = """Variation of no. of Susceptible, infected and Removed people with time 
0.3 Infection Radius
100 Populaton 
3% initial infected
"""
title_arr = ["Infection Duration = 60", "Infection Duration = 120", "Infection Duration = 180", "Infection Duration = 240"]
data_arr = [dataf03_60_100_3, dataf03_120_100_3, dataf03_180_100_3, dataf03_240_100_3]
subplotGraph2D(2, 2, data_arr, title_arr, title)
```
