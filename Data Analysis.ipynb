```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("data.csv", header = None)
data.columns = ["S", "I", "R"]
data.head()
```

```python
data.info()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(y = data["S"], mode = "lines", name = "SUSCEPTIBLE"))
fig.add_trace(go.Scatter(y = data["I"], mode = "lines", name = "INFECTED"))
fig.add_trace(go.Scatter(y = data["R"], mode = "lines", name = "RECOVERED OR REMOVED"))
fig.show()
```
